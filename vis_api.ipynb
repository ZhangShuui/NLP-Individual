```python
# 读取test_results/中的所有文件
import os
import json
import re
import pandas as pd
import copy
def read_files_in_directory(directory):
    # 获取目录下所有文件
    files = os.listdir(directory)
    data = []

    for file in files:
        if file.endswith('.log'):
            file_name = copy.copy(file)
            # file_name的格式为：{model_name}_{prompt_name}_{dataset}.log
            model_name = file.split('_')[0]
            file = file.replace(model_name + '_', '')
            if "dev_matched" in file:
                dataset = "dev_matched"
            elif "dev_mismatched" in file:
                dataset = "dev_mismatched"
            prompt_name = file.replace('.log', '').replace("_" + dataset, '')
            with open(os.path.join(directory, file_name), 'r') as f:
                content = f.readlines()
                # 提取log文件中的Accuracy和F1值
                # === Evaluation Results ===
                # accuracy: 0.8000
                # macro_f1: 0.7222
                for line in content:
                    if 'accuracy' in line:
                        accuracy = re.search(r'accuracy: ([\d.]+)', line)
                        if accuracy:
                            accuracy = float(accuracy.group(1))
                    elif 'macro_f1' in line:
                        f1 = re.search(r'macro_f1: ([\d.]+)', line)
                        if f1:
                            f1 = float(f1.group(1))
                # 将数据存储到字典中
                result = {
                    'model_name': model_name,
                    'prompt_name': prompt_name,
                    'dataset': dataset,
                    'accuracy': accuracy,
                    'f1': f1
                }
                data.append(result)                            
                            
    return data

def save_to_csv(data, output_file):
    # 将数据转换为DataFrame
    df = pd.DataFrame(data)
    # 保存为CSV文件
    # df.to_csv(output_file, index=False)
    # print(f"Data saved to {output_file}")
    return df

data = read_files_in_directory('test_results/api_models/')
data
```

```python
data = save_to_csv(data, 'test_results.csv')
```
